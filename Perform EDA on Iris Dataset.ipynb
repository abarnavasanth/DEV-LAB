```python
# Histograms of selected features 
features = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm'] 
for feature in features: 
  plt.figure (figsize = (6, 4)) 
sns.histplot (data [feature], kde=True, bins=20) 
plt.title (f"Distribution of {feature}") 
plt.show ( )
```

```python
# Pairplot for distributions and relationships
sns.pairplot(data, hue="Species", palette="Set2", diag_kind="kde")
plt.show()
```

```python
# Calculate correlation matrix for numerical features
correlation_matrix = data[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']].corr()

# Display the correlation matrix
print(correlation_matrix)
```

```python
# Plotting the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Iris Features')
plt.show()
```

```python
# Violin plot to show distribution and density of the data based on Species
plt.figure(figsize=(10, 6))

for i, feature in enumerate(['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']):
    plt.subplot(2, 2, i + 1)
    sns.violinplot(x='Species', y=feature, data=data, palette='Set2')
    plt.title(f'Violin Plot of {feature}')
    plt.tight_layout()

plt.show()
```

```python
# Boxplots for each feature based on Species
plt.figure(figsize=(10, 6))

for i, feature in enumerate(['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']):
    plt.subplot(2, 2, i + 1)
    sns.boxplot(x='Species', y=feature, data=data, palette='Set2')
    plt.title(f'Boxplot of {feature}')
    plt.tight_layout()

plt.show()
```
